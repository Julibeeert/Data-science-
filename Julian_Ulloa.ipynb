```python
# Instalar yfinance
# !pip install yfinance

import yfinance as yf

# Descargar datos históricos de Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")  # Puedes usar "5y", "1y", etc.

# Restablecer el índice para tener la fecha como columna
tesla_data.reset_index(inplace=True)

# Visualizar las 5 primeras filas
tesla_data.head()
```

```python
import pandas as pd
import requests

# URL de la página de ingresos de Tesla
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Simular navegador con headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36"
}

# Obtener contenido HTML con headers
response = requests.get(url, headers=headers)
html = response.text

# Extraer las tablas del HTML
tables = pd.read_html(html)

# Mostrar cuántas tablas se encontraron
print(f"Se encontraron {len(tables)} tablas")

# Revisar las primeras filas de cada tabla (opcional para encontrar la correcta)
for i, table in enumerate(tables):
    print(f"\nTabla {i}:\n")
    print(table.head())

# Elegir la tabla que tiene la columna 'Revenue'
tesla_revenue = tables[1]  # Cámbiala si no es esta

# Mostrar las 5 últimas filas
tesla_revenue.tail()
```

```python
import matplotlib.pyplot as plt

# Función para graficar
def make_graph(data, title):
    plt.figure(figsize=(14, 6))
    plt.plot(data['Date'], data['Close'], label='Cierre')
    plt.xlabel('Fecha')
    plt.ylabel('Precio de Cierre (USD)')
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Usar la función con los datos de Tesla
make_graph(tesla_data, "Tesla Stock Price Over Time")
```
